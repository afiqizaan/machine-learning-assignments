##### Dataset

```python
# Importing wine dataset
import numpy as np
import pandas as pd

data = pd.read_csv("/content/wine.csv")
```

##### Data Preprocessing

```python
# Scaling the dataset using StandardScaler
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaled_data = scaler.fit_transform(data)

scaled_data = pd.DataFrame(scaled_data, columns=data.columns)
scaled_data.describe()
```

##### EDA

```python
# Finding high correlation; decided to not drop any features
import matplotlib.pyplot as plt
import seaborn as sns
corr = scaled_data.corr()
plt.figure(figsize=(9,6))
sns.heatmap(corr, annot=True, fmt='.2f', cmap='coolwarm')
plt.show()
```

```python
# Finding outliers
columns = ["Alcohol", "Malic_Acid", "Ash", "Magnesium", "Total_Phenols",
           "Flavanoids", "Nonflavanoid_Phenols", "Proanthocyanins",
           "Color_Intensity", "Hue", "OD280", "Proline"]

fig, axes = plt.subplots(3, 4, figsize=(18, 7))

for i, col in enumerate(columns):
  ax = axes[i // 4][i % 4]
  ax.hist(scaled_data[col], bins=10, edgecolor="black")
  ax.set_title(col)
  ax.set_xlabel("Values")
  ax.set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

```python
# Removing outliers
for column in scaled_data.columns:
  scaled_data = scaled_data[~((scaled_data[column] - scaled_data[column].mean()).abs() > 3 * scaled_data[column].std())]
```

```python
# Re-checking if there are any more outliers
columns = ["Alcohol", "Malic_Acid", "Ash", "Magnesium", "Total_Phenols",
           "Flavanoids", "Nonflavanoid_Phenols", "Proanthocyanins",
           "Color_Intensity", "Hue", "OD280", "Proline"]

fig, axes = plt.subplots(3, 4, figsize=(17, 8))

for i, col in enumerate(columns):
  ax = axes[i // 4][i % 4]
  ax.hist(scaled_data[col], bins=10, edgecolor="black")
  ax.set_title(col)
  ax.set_xlabel("Values")
  ax.set_ylabel("Frequency")

plt.tight_layout()
plt.show()
```

##### KMeans (PCA)

```python
# Reducing the dimensionality of the dataset using PCA
from sklearn.decomposition import PCA
pca = PCA(n_components=2)
reduced_data = pca.fit_transform(scaled_data)
```

```python
# Using elbow method to find suitable n_clusters
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia_scores = []
cluster_num = []

for n in range(1, 10):
  model = KMeans(n_clusters=n, init='k-means++', random_state=42)
  model.fit(reduced_data)
  inertia_scores.append(model.inertia_)
  cluster_num.append(n)

plt.plot(cluster_num, inertia_scores, 'o-')
plt.xlabel("Number of Clusters")
plt.ylabel("Inertia")
plt.show()
```

```python
# Deploying KMeans model with n_clusters=3
model = KMeans(n_clusters=3, init='k-means++', random_state=42)
model.fit(reduced_data)
labels = model.predict(reduced_data)
centers = model.cluster_centers_
labels = model.labels_
```

```python
# Plotting the clusters with scatterplot
plt.scatter(reduced_data[:, 0], reduced_data[:, 1], c=labels, cmap='viridis', marker='o')
plt.scatter(centers[:, 0], centers[:, 1], marker='x', c='black')
plt.title('K-means Clustering (PCA)')
plt.xlabel('PCA Feature 1')
plt.ylabel('PCA Feature 2')
plt.show()
```

```python
# Printing silhouette score (PCA)
from sklearn.metrics import silhouette_score
silhouette_score_pca = silhouette_score(reduced_data, labels)
print(f"Silhouette score: {silhouette_score_pca}")
```

##### KMeans (Autoencoder)

```python
# Using Autoencoder with set seed 42
import tensorflow as tf
from tensorflow.keras.models import Model
from tensorflow.keras.layers import Input, Dense
from tensorflow.keras.optimizers import Adam

input_layer = Input(shape=(scaled_data.shape[1],))
encoded = Dense(2, activation='relu')(input_layer)
decoded = Dense(scaled_data.shape[1], activation='sigmoid')(encoded)
autoencoder = Model(inputs=input_layer, outputs=decoded)
encoder = Model(inputs=input_layer, outputs=encoded)

autoencoder.compile(optimizer=Adam(), loss='mean_squared_error')
autoencoder.fit(scaled_data, scaled_data, epochs=50, batch_size=32, validation_split=0.2)
```

```python
# Deplyong KMeans model with n_clusters=3
reduced_data_ae = encoder.predict(scaled_data)
model_ae = KMeans(n_clusters=3, init='k-means++', random_state=42)
labels_ae = model_ae.fit_predict(reduced_data_ae)
centers_ae = model_ae.cluster_centers_
```

```python
# Plotting the clusters with scatterplot
plt.scatter(reduced_data_ae[:, 0], reduced_data_ae[:, 1], c=labels_ae, cmap='viridis', marker='o')
plt.scatter(centers_ae[:, 0], centers_ae[:, 1], marker='x', c='black')
plt.title('K-means Clustering (Autoencoder)')
plt.xlabel('Autoencoder Feature 1')
plt.ylabel('Autoencoder Feature 2')
plt.show()
```

```python
# Printing silhouette score (Autoencoder)
silhouette_score_ae = silhouette_score(reduced_data_ae, labels_ae)
print(f"Autoencoder Silhouette score: {silhouette_score_ae}")
```

##### PCA vs Autoencoder

```python
# Printing both silhouette scores
print(silhouette_score_pca)
print(silhouette_score_ae)
```

```python
# Comparing both silhouette scores
silhouette = ["Silhouette Score (PCA)", "Silhouette Score (AE)"]
plt.bar(silhouette[0], silhouette_score_pca)
plt.bar(silhouette[1], silhouette_score_ae)
plt.title("PCA vs Autoencoder")
plt.show()
```
